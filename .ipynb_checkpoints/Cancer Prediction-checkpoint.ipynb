# Preparing Data before Trainning

```python
# WE SELECT BIOPSY DATA AS OUR TARGET VALUES:

target_df = cancer_df['Biopsy']
input_df = cancer_df.drop(['Biopsy'], axis=1)
```

```python
X = np.array(input_df).astype('float32')
y = np.array(target_df).astype('float32')

y = y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
# SPLITTING DATA INTO TRAIN AND TEST DATASETS
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5)
```

# Trainning & Evaluation 

```python
import xgboost as xgb
model = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 50, n_estimators = 100)
model.fit(X_train, y_train)
```

```python
result_train = model.score(X_train, y_train)
result_train
```

```python
result_test = model.score(X_test, y_test)
result_test
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_predict))
```

Classification Report:
* This report presents the performance of a binary classifier. The metrics displayed include precision, recall, F1-score, and support for each class (0 and 1).
* Precision: The ratio of correctly predicted positive observations to the total predicted positives. It is 0.94 for class 0 and 1.00 for class 1.
Recall (Sensitivity): The ratio of correctly predicted positive observations to all observations in the actual class. It is 0.97 for class 0 and 1.00 for class 1.
* F1-score: The weighted average of Precision and Recall. It takes both false positives and false negatives into account. It is 0.97 for class 0 and 0.55 for class 1.
* Support: The number of actual occurrences of the class in the dataset. Class 0 has 83 instances and class 1 has 3 instances.
* Additional metrics like accuracy, macro average, and weighted average are also provided. The overall accuracy of the model on the test set is 0.94, which means it correctly predicted 94% of the class labels.

```python
cm = confusion_matrix(y_predict, y_test)
sns.heatmap(cm, annot = True)
plt.show()
```

The confusion matrix for the classification is shown as a heatmap. The matrix layout is:
* Top Left (True Negative): 78 - The model correctly predicted 78 instances of class 0 as class 0.
* Top Right (False Positive): 0 - The model made no incorrect predictions of class 1 as class 0.
* Bottom Left (False Negative): 5 - The model incorrectly predicted 5 instances of class 0 as class 1.
* Bottom Right (True Positive): 3 - The model correctly predicted 3 instances of class 1 as class 1.
